````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_json("coupler-NCap-cap_matrix.json")
````

````python
def flatten(d, parent_key="", out=None):
    if out is None:
        out = {}

    for k, v in d.items():
        key = f"{parent_key}.{k}" if parent_key else k
        if isinstance(v, dict):
            flatten(v, key, out)
        else:
            out[key] = v
    return out
````

````python

# in the coupler-NCap-cap_matrix there are different size design dictionaries for some reason
total_dict = {}
for newkey in flatten(df.design.iloc[0]):
    total_dict[newkey] = []


for i in df.design:
    this_dict = flatten(i)
    for k in this_dict:
        try:
            total_dict[k] += [this_dict[k]]
        except:
            total_dict[k] = [this_dict[k]]
````

````python
multi_valued_params = []
single_valued_params = []

for j in total_dict:
    uniques = np.unique(total_dict[j])
    print("Parameter {}".format(j))
    print("Number of unique values:",len(uniques))
    print("Possible unique value(s):",uniques)
    print()

    if len(uniques) > 1:
        multi_valued_params.append(j)
    else:
        single_valued_params.append(j)
````
