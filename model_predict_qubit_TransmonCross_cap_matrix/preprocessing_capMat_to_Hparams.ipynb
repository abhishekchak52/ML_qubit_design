```python
from squadds import SQuADDS_DB
import pandas as pd
from squadds import Analyzer
from squadds import AnsysSimulator
import numpy as np
import json
```

```python
df = pd.read_json("qubit-TransmonCross-cap_matrix.json")
```

```python
db = SQuADDS_DB()
db.select_system("qubit")
db.select_qubit("TransmonCross")
analyzer = Analyzer(db)
```

Below we make a "faux" AnsysSimulator object so we can use the Hamiltonian calculation features

```python
ansys_simulator = AnsysSimulator(analyzer,None)
```

```python
%%capture
Hamiltonian_params = []
for index in df.index:
    device_dictionary = {"design": df[df.index == index].design.iloc[0],
                      "sim_options": df[df.index == index].sim_options.iloc[0],
                      "sim_results": df[df.index == index].sim_results.iloc[0]}
    
    Hamiltonian_params.append(ansys_simulator.get_xmon_info(device_dictionary))
```

```python
df["Hamiltonian_params"] = Hamiltonian_params
```

```python
df.to_json("qubit-TransmonCross-Hamiltonian_params.json",orient="records",indent=3)
```
